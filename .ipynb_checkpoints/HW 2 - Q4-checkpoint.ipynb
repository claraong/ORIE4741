# (a)

The graph of $f(X + \alpha v)$ is always the tangent to the graph of $f(X) + \alpha (\nabla f)^T v$ at $\alpha = 0$. The graphs intersect when alpha is zero.

```julia
# generate random data matrix
srand(3)
v = randn(5)
X = randn(5)
```

```julia
f(X) = log(sum(exp(X)))
```

```julia
using PyPlot
```

```julia
alphas = linspace(-1,1,100) #100 samples to generate
∇f = exp(X)/sum(exp(X))
plot(alphas, [f(X) + alpha*dot(∇f, v) for alpha in alphas], label=L"$f(X) + \alpha (\nabla f)^T v$")
plot(alphas, [f(X + alpha*v) for alpha in alphas], label=L"$f(X + \alpha v)$") #latex string
xlabel(L"$\alpha$")
legend()
```
