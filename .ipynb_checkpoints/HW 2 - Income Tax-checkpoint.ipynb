# b)

Continuing to look only at Tompkins County, fit a model that predicts avg_tax using the year number. Do this with least squares.

```julia
function plotline(w;
                  xmin=-100,xmax=100)
    xsamples = [xmin, xmax]
    plot(xsamples, [w*x for x in xsamples], color="black")
end
```

```julia
X =array(tompkins_total[:,:Year])
Y = array(tompkins_total[:,:avg_tax])
w = X\Y
println(w) # 0.00106
```

Transform the input to include an offset term in the model. Call the coefficients from this model $w^b$.

```julia
X2 = [copy(X) ones(length(X))]
w2=X2\Y
println(w2) # [0.05272998568828175,-103.64804933914492]
```

# c)

Now we will change the feature space.
Fit a model using the year and the avg_tax from the previous year.

```julia
lengthY = size(Y)[1]
lengthX = size(X)[1]
prev_avgtax = Y[1:lengthY-1,:] # 14 by 1
years = X[2:lengthX,:] #14 by 1
Y3 = Y[2:lengthY,:] # 14 by 1
X3= [years prev_avgtax]
w3 = X3\Y3
println(w3) # [0.00021548417733040904  0.8169915269988163]
```

Call these coefficients $w^c$. Interpret these coefficients. What do they mean?


The average tax rate is linearly proportional to the year and the average tax from the previous year. Over time, the average tax rate increases. As the average tax increases, the average tax for the next year also increases. Note that this shows correlation and not causality.

Plot the real average tax return per year against your model's prediction. Discuss how you think your model fits.



```julia
predicted = X3*w3
Gadfly.plot(x=Y3, y=predicted, Guide.xlabel("Actual"), Guide.ylabel ("Predicted"), Geom.point)
```

```julia
error = abs2(predicted.-Y3)

mse = sum(error[:,1])/size(error)[1] # 0.0134
println(mse)
```

The predicted values tend to be slightly higher than the actual values. The model fit can be improved by adding more features.

Add two new features to your model. This could be a column of the data, a transformation of a column, or a new column formed from another. The only requirement is that the prediction in year $t$ be based on data avilable by the end of year $t-1$

Fit this model and interpret the coefficients. Call these $w^d$ 

```julia
lengthY = size(Y)[1]
prev_avgtax = Y[3:lengthY-1,:] # 12 by 1 # one year ago
prev2_avgtax = Y[2:lengthY-2,:] # 12 by 1 # two years ago
prev3_avgtax = Y[1: lengthY-3,:] # 12 by 1 # three years ago
Y4 = Y[4:lengthY,:] # 12 by 1
X4 = [X[4:size(X)[1],:] prev_avgtax prev2_avgtax prev3_avgtax] #12 by 1
w4 = X4\Y4
println(w4) # [0.000183611949747483 0.9672078314743522 -0.681218670066828 0.5777305629200722]
```

```julia
actual = Y[4: lengthY,:]
predicted = X4*w4
Gadfly.plot(x=actual, y = predicted, Guide.xlabel ("Actual"), Guide.ylabel("Predicted"))
```

```julia
error = abs2(predicted.-actual)

mse = sum(error[:,1])/ size(error)[1] # 0.004119
println(mse)
```

Feature space: Avg tax from 1 year before, avg tax from 2 years before and avg tax from 3 years before. Features fit the data well as the insample error is small at 8%.

Now we want to see how this model performs in other counties. Apply the model you fit in part c, with coefficients $w^c$, to data from the other counties and compute the error of this model on each data set.

```julia
# only select the total income values
tax_total = tax[tax[:IncomeClass] .=="Total",:]

println(size(tax_total)) # 973 by 14
```

```julia
# remove fake counties
fake_counties = String[]
push!(fake_counties,"NYS Unclassified +")
push!(fake_counties,"Residence Unknown ++")
push!(fake_counties, "Grand Total, Full-Year Resident")
relevant_counties = Array (Bool, size(tax_total,1))

for (i,county) in enumerate(tax_total[:County]) # i is the index
    relevant_counties[i] = ~(county in fake_counties) 
end
#relevant_counties
#println(sum(relevant_counties)) #929
```

```julia
# remove counties which did not report every year
relevant_tax = tax_total[relevant_counties,:]

counties = Set((relevant_tax[:County])) # No duplicates. 62 unique counties.

reporting_counties = String[]

for county in counties
    if size(relevant_tax[relevant_tax[:County].==county,:County])[1] == 15 # 15 years of records
        push!(reporting_counties, county)
    end
end
#println(size(reporting_counties,1)) # 61. Only Greene did not report for all years.

reported_tax_all_years = Array(Bool, size(relevant_tax,1)) 
for (i,county) in enumerate(relevant_tax[:County]) # i is the index
    reported_tax_all_years[i] = (county in reporting_counties)
end

cleaned_tax = relevant_tax[reported_tax_all_years,:] # 915 entries
```

# g)

Using the same features you chose in part d, fit a model to the data for each of the other counties. These will be county-specific models. They may have different coefficient vectors from $w^d$.

Plot a histogram of the mean square errors of the county-specific models for their respective counties.

```julia
errors2 = Array {Float64, 1}

errors2 = zeros(61)
i = 1

for county in reporting_counties # unique counties
    data = cleaned_tax[[cleaned_tax[:County] .== county],:] # 15
    length = size(data)[1]
    prev_avgtax = data[3:(length-1),:avg_tax]
    prev2_avgtax = data[2:(length-2),:avg_tax] 
    prev3_avgtax = data[1: (length-3),:avg_tax] # 12 by 1 # three years ago
    
    years = data[4:size(data)[1],:Year] 
    X7 = array([years prev_avgtax prev2_avgtax prev3_avgtax])
    
    Y7 = array(data[4:size(data)[1],:avg_tax])
    w7 = X7\Y7    
    predicted = X7*w7 # using county-specific models
    error = abs2(predicted.-Y7)
    mse = sum(error[:,1])/size(error,1) # 
    errors2[i] = mse
    i += 1
end
```

```julia
# plot histogram
plt[:hist](errors2, 10)
println("Median:", median(errors2), " Mean:", mean(errors2))
```

Again, are there major outliers? How do the coefficients compare across counties?

Error in Tompkins: 0.004119. Most of the errors (37 out of 61) in other datasets are smaller than 0.004119. Westchester has a high error of 0.543 and NYC Manhattan has an even higher of 1.25. These are outliers. The coefficients are different across counties.

```julia
# remove outliers
removed_outliers = errors2
errors_revised = deleteat! (removed_outliers, [29 59]) # index position of outliers: 29 and 59
plt[:hist](errors_revised, 20)
println("Median:", median(errors_revised), " Mean:", mean(removed_outliers))
```

Removing the outliers, the mean of the mean squared errors (0.00680) is close to 0.004119, the error for Tompkins. This means that the Tompkins model generalizes well to other counties in most cases.
