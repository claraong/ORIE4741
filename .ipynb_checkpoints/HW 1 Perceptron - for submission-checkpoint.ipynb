# Generate data
Generate 50 random datapoints in 2 dimensions.
Note that the last column is fixed at 1 to capture the offset

```julia
# don't change this: it sets the random number generator, 
# so everyone in the class will have consistent results
srand(17)

# set dimensions
n = 50
d = 2
# generate input: the rows of X are the data points
# notice the last column encodes the offset
X = [randn(n, d) ones(n)]

# Generate the target function w♮
# Store the correct decision for each datapoint in y
w♮ = randn(d+1)
y = sign(X*w♮);
```

```julia
# plot data points X and color according to y
clf()
hold(true)
axis("off")
for i=1:n
    plot(X[i,1], X[i,2], y[i]==1 ? "ro" : "bd")
end

# set figure limits
ylim([-2,2])
xlim([-2,2])
savefig("perceptron_data_hw1.eps")
```

```julia

"""
This function takes data points X, plots them, and colors them correctly according to y.
This function also plots the line w'*x = 0.
We are plotting just the first two dimensions of x; 
the last (offset) coordinate of w sets the offset of the line from the origin.
"""
function plot_perceptron(X,y,w)
    # plot data points X and color according to y
    clf()
    hold(true)
    axis("off")
    for i=1:n
        plot(X[i,1], X[i,2], y[i]==1 ? "ro" : "bd")
    end
    
    # plot vector w
    x1samples = [minimum(X[:,1]), maximum(X[:,1])] # first dimension
    
    if w[2]!=0
        plot(x1samples, [-(w[1]*x1 + w[3])/w[2] for x1 in x1samples], color="black")
    end
    
    # set figure limits
    ylim([-2,2])
    xlim([-2,2])
    println(w)
end
```

```julia
# Double check our function by plotting the points with the truth vector 
plot_perceptron(X,y,w♮)
```

# a) Code the missing parts of the Perceptron algorithm
Inputs to Perceptron are: 
* required arguments:
    * X = data points
    * Y = true values
* optional arguments:
    * maxsteps = maximum number of updates to w (default: 100)
    * w = initial guess of w (default: random)

Perceptron should start with the input guess of w, make at most maxsteps updates, and return the final w.

```julia
function perceptron(X,y;
                    maxsteps=200, # 
                    w = rand(size(X,2))) # initial guess for w. Creates 3 random numbers.
    
    ## YOUR CODE HERE ##
    solution_reached = false
    numUpdates = 0
    num_row = size(X, 1)
    
    
    shuffle_array = shuffle(collect(1:num_row)) #returns a random permutation of numbers from 1 to 50. For part e.
    
    while !solution_reached 
        solution_reached = true
        pred_y = sign(X*w)
        
        for i in shuffle_array # go through every data point in X
            if pred_y[i] != y[i] && numUpdates < maxsteps
                w += (y[i] * X[i,:])' # w will be a vertical vector
                numUpdates += 1 # number of times w updates
                solution_reached = false # misclassified
                break
            end
        end
    end
    
    return w, numUpdates
end
```

# b) Run perceptron on linearly separable data set where n = 50

```julia
srand(17)
n = 50
d = 2
X = [randn(n, d) ones(n)]
y = sign(X*w♮);
w_final = perceptron(X,y; maxsteps=1000)
plot_perceptron(X,y,w_final[1])
@time(perceptron(X,y))
println("Number of updates: ", w_final[2]) 
```

# h) Try out one idea to fix perceptron

```julia
function perceptron_revised (X,y;
                    maxsteps=200, # 
                    w = rand(size(X,2))) # initial guess for w. Creates 3 random numbers.
    
    ## YOUR CODE HERE ##
    solution_reached = false
    numUpdates = 0
    num_row = size(X, 1)
    
    
    shuffle_array = shuffle(collect(1:num_row)) #returns a random permutation of numbers from 1 to 50. For part e.
    
    while !solution_reached 
        solution_reached = true
        pred_y = sign(X*w)
        
        for i in shuffle_array
            increment = (y[i] * X[i,:])'

            increment_magnitude = ((increment)'*increment)[1]
         
            if pred_y[i] != y[i] && numUpdates < maxsteps && increment_magnitude < 1.8 # limit the size of the increment
                w += increment  # w will be a vertical vector
                solution_reached = false # misclassified
                numUpdates += 1
                break
            end
        end
        
    end
    
    return w, numUpdates
end

perceptron_revised(X,y,maxsteps=50)
plot_perceptron(X, y, perceptron(X,y; maxsteps=10000)[1])
println("Number of updates: ", perceptron(X,y)[2])

```
